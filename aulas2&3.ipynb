```python
import pandas as pd
import plotly.express as px
```

```python
df_principal = pd.read_excel("/content/Imersão Python - Tabela de ações - ORIGINAL.xlsx", sheet_name="Principal")
df_principal
```

```python
df_total_acoes = pd.read_excel("/content/Imersão Python - Tabela de ações - ORIGINAL.xlsx", sheet_name="Total_de_acoes")
df_total_acoes
```

```python
df_ticker = pd.read_excel("/content/Imersão Python - Tabela de ações - ORIGINAL.xlsx", sheet_name="Ticker")
df_ticker
```

```python
df_principal = df_principal[['Ativo',	'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

```python
df_principal = df_principal.rename(columns={'Último (R$)':'valor_final', 'Var. Dia (%)':'var_dias_pct'}).copy()
df_principal
```

```python
df_principal['Var_pct'] = df_principal['var_dias_pct'] / 100
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['Var_pct'] + 1)
df_principal
```

```python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Código'])
df_principal
```

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['Qtde. Teórica']
df_principal
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal
```

```python
df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)
df_principal
```

```python
df_principal = df_principal.rename(columns={'Qtde. Teórica':'qtde_teorica'}).copy()
df_principal
```

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
```

```python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
df_principal
```

```python
df_principal = df_principal.drop(columns=['Ticker'])
df_principal
```

```python
df_chatgpt = pd.read_excel("/content/Imersão Python - Tabela de ações - ORIGINAL (1).xlsx", sheet_name='Chatgpt')
df_chatgpt
```

```python
df_principal = df_principal.merge(df_chatgpt, left_on='Nome', right_on='Nome empresa', how='left')
df_principal = df_principal.drop(columns=['Nome empresa'])
df_principal
```

```python
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade'})
df_principal
```

```python
df_principal['cat_idade'] = df_principal['Idade'].apply(lambda x: 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
df_principal
```

```python
maior = df_principal['variacao_rs'].max()
menor = df_principal['variacao_rs'].min()
media = df_principal['variacao_rs'].mean()
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Media\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
```

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

```python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_rs', text='variacao_rs', title='Variação Reais x Resultado')
fig.update_traces(texttemplate='R$ %{text:.2f}', textposition='outside')
fig.show()
```

```python
fig2 = px.pie(df_analise_segmento, values='variacao_rs', names='Segmento', title='Variação x Segmento')
fig2.show()
```

```python
df_analise_faixa_etaria = df_principal.groupby('cat_idade').agg({'variacao_rs':'sum', 'cat_idade':'count'})
df_analise_faixa_etaria
```

```python
df_analise_faixa_etaria = df_analise_faixa_etaria.rename(columns={'cat_idade':'qtd'})
```

```python
df_analise_faixa_etaria.reset_index(inplace=True)
df_analise_faixa_etaria
```

```python
fig3 = px.bar(df_analise_faixa_etaria, y='cat_idade', x='variacao_rs', title='Variação Reais x Faixa Etaria', orientation='h')
fig3.show()
```

```python
fig4 = px.bar(df_analise_faixa_etaria, x='cat_idade', y='qtd', title='Quantidade por Faixa Etaria', orientation='v')
fig4.show()
```
